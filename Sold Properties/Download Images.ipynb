```python
import pandas as pd
import numpy as np
import os
import requests
from bs4 import BeautifulSoup
import re

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

from tqdm import tqdm  # Import tqdm for progress bar
```

```python
# Define the directory where the CSV files are located
directory = './'  # Update this with the appropriate directory path

# Function to process each CSV file
def process_csv(file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Remove duplicates based on "List ID" column, keeping the oldest "Sold Date"
    df['Sold Date'] = pd.to_datetime(df['Sold Date'])  # Convert 'Sold Date' to datetime
    df.sort_values(by='Sold Date', inplace=True)  # Sort by 'Sold Date' to get the oldest first
    df.drop_duplicates(subset='List ID', keep='first', inplace=True)
    
    return df

# List to store processed DataFrames
dfs = []

# Iterate through files in the directory
for file in os.listdir(directory):
    # Check if the file is a CSV file and starts with "Sold Properties"
    if file.endswith('.csv') and file.startswith('Sold Properties'):
        # Process the CSV file and append the DataFrame to the list
        file_path = os.path.join(directory, file)
        df = process_csv(file_path)
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dfs, ignore_index=True)
```

```python
image_df = final_df[['Title', 'Href', 'Sold Date', 'List ID']]
```

```python
# Function to extract image links
def extract_image_links(url):
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        image_links = []
        # Find all <a> tags with class 'gridImage'
        for a_tag in soup.find_all('a', class_='gridImage'):
            # Extract the 'data-background' attribute
            data_background = a_tag.get('data-background')
            # Check if it's a .jpg link and append to the list
            if data_background and data_background.endswith('.jpg'):
                image_links.append(data_background)
        return image_links
    else:
        return []
```

```python
# Apply function to extract image links and create new column
tqdm.pandas()  # Use tqdm for progress bar
image_df['Image Links'] = image_df['Href'].progress_apply(extract_image_links)
```

```python
image_df = image_df.explode('Image Links')
```

```python
image_df['Image Name'] = image_df['List ID'] + '_' + (image_df.groupby('List ID').cumcount() + 1).astype(str)

image_df
```

```python
# Folder location
folder_path = "C:\\Users\\lakha\\OneDrive\\Documents\\House Flipping - Real Life\\Sold House Images"

# Save images
for index, row in image_df.iterrows():
    list_id = row['List ID']
    image_name = row['Image Name']
    image_link = row['Image Links']

    if pd.notnull(image_link):  # Check if the image link is not NaN
        try:
            # Create directories based on List ID
            list_folder = os.path.join(folder_path, list_id)
            if not os.path.exists(list_folder):
                os.makedirs(list_folder)
            
            # Download the image using requests
            response = requests.get(image_link)
            if response.status_code == 200:
                image_path = os.path.join(list_folder, f"{image_name}.jpg")
                with open(image_path, 'wb') as f:
                    f.write(response.content)
                print(f"Image saved: {image_path}")
            else:
                print(f"Failed to download image: {image_link}")
        except Exception as e:
            print(f"Error downloading image: {str(e)}")
    else:
        print(f"Skipping NaN image link for {image_name}")
```
